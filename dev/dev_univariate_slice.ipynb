```python
from mfusampler.main import uni_slice
import numpy as np

def log_density(x):
    return -0.5 * x**2

x_init = 0.0
n_sample = 10000
x = x_init
samples = []
for n in range(n_sample):
    x = uni_slice(x, log_density, w = 1, m = 10)
    samples.append(x)
samples = np.array(samples)
```

```python
import matplotlib.pyplot as plt

indices = np.arange(len(samples))

plt.figure(figsize=(10, 5))

plt.plot(indices, samples)
plt.xlabel('Iteration No')
plt.ylabel('Sample')
plt.title('MCMC Samples')

#plt.legend()

# Show the plot
plt.show()
```

```python
plt.hist(samples)
plt.xlabel('x')
plt.ylabel('Frequency')
plt.title('Histogram of Samples')

plt.show()
```
